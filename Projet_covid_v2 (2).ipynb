```python
def clean_test(X): #Comme chiffre en format fr, clean les ". puis retirer les ","" et remplacer par les "."
    X = str(X)
    X = X.replace(".", "")
    X = X.replace(",", ".")
    X = X.replace("''", "")
    X = np.float32(X)
    #df_clean= df_clean.astype(float)
    return X
```

```python
def import_total_mensuel(file_name):
  df_test = pd.DataFrame(pd.read_csv(f"{file_name}.csv", sep=";",index_col="Type_depenses" ))
  df_test = df_test.T
  return df_test
```

```python
from pandas import read_csv
from pandas import to_datetime
from pandas import DataFrame
from fbprophet import Prophet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from matplotlib import pyplot
from google.colab import files
```

```python
def train_fit_prophete(col_target,year_pred):
	# prepare expected column names
	full_dataframe_transform = full_dataframe.reset_index().rename(columns={'index':'ds', f"{col_target}":'y'})
	# create test dataset, remove last 24 months 2020-2021
	train = full_dataframe_transform.drop(full_dataframe_transform.index[(year_pred-2020)*12-24:])
	#print(train.tail())
	# define the model
	model = Prophet()
	# fit the model
	model.fit(train)
	# define the period for which we want a prediction
	future = list()
	for i in range(1, 13):
		date = f"{year_pred}-{i}-01"
		future.append([date])
	future = DataFrame(future)
	future.columns = ['ds']
	future['ds'] = to_datetime(future['ds'],format="%Y/%m/%d")
	# use the model to make a forecast
	forecast = model.predict(future)
	# calculate MAE between expected and predicted values for 2020
	y_true = full_dataframe_transform['y'][(year_pred-2020)*12-24:(year_pred-2020)*12-12].values
	y_pred = forecast['yhat'].values
	mape = mean_absolute_percentage_error(y_true, y_pred)
	print('MAPE: %.3f' % mape)
	# plot expected vs actual
	pyplot.plot(y_true, label='Actual')
	pyplot.plot(y_pred, label='Predicted')
	pyplot.legend()
	pyplot.show()
```

```python
def train_fit_prophete_complete(col_target):
	# prepare expected column names
	full_dataframe_transform = full_dataframe.reset_index().rename(columns={'index':'ds', f"{col_target}":'y'})
	# create test dataset, remove last 36 months 2019-2021
	train = full_dataframe_transform.drop(full_dataframe_transform.index[-36:])
	#print(train.tail())
	# define the model
	model = Prophet()
	# fit the model
	model.fit(train)
 
	# define the period for which we want a prediction 
	future = list()
	for i in range(1, 13):
		date = f"{2020}-{i}-01"
		future.append([date])
	future = DataFrame(future)
	future.columns = ['ds']
	future['ds'] = to_datetime(future['ds'],format="%Y/%m/%d")
	# use the model to make a forecast
	forecast = model.predict(future)
 
	# calculate MAE between expected and predicted values for 2020
	y_true = full_dataframe_transform['y'][-24:-12].values
	y_pred = forecast['yhat'].values
	mape = mean_absolute_percentage_error(y_true, y_pred)
	mae = mean_absolute_error(y_true, y_pred)
	print('MAE: %.3f' % mae)
	print('MAPE: %.3f' % mape)
	# plot expected vs actual
	pyplot.plot(y_true, label='Actual')
	pyplot.plot(y_pred, label='Predicted')
	pyplot.legend()
	pyplot.show()
 
 	# calculate MAE between expected and predicted values for 2019
	y_test = full_dataframe_transform['y'][-36:-24].values
	y_pred = forecast['yhat'].values
	mape = mean_absolute_percentage_error(y_test, y_pred)
	mae = mean_absolute_error(y_test, y_pred)
	print('MAE: %.3f' % mae)
	print('MAPE: %.3f' % mape)
	# plot expected vs actual
	pyplot.plot(y_test, label='Actual')
	pyplot.plot(y_pred, label='Predicted')
	pyplot.legend()
	pyplot.show()
```

```python

def train_fit_prophete_complete_2019_2021(col_target):
	# prepare expected column names
	full_dataframe_transform = full_dataframe.reset_index().rename(columns={'index':'ds', f"{col_target}":'y'})
	# create test dataset, remove last 36 months 2019-2021
	train = full_dataframe_transform.drop(full_dataframe_transform.index[-36:])
	#print(train.tail())
	# define the model
	model = Prophet()
	# fit the model
	model.fit(train)

 	# define the period for which we want a prediction for 2021
	future = list()
	for i in range(1, 13):
		date = f"{2021}-{i}-01"
		future.append([date])
	future = DataFrame(future)
	future.columns = ['ds']
	future['ds'] = to_datetime(future['ds'],format="%Y/%m/%d")
	# use the model to make a forecast
	forecast = model.predict(future)

	# calculate MAE between expected and predicted values for 2021
	y_true_2021 = full_dataframe_transform['y'][-12:].values
	y_pred_2021 = forecast['yhat'].values
	 
	mape_2021 = mean_absolute_percentage_error(y_true_2021, y_pred_2021)
	mae_2021 = mean_absolute_error(y_true_2021, y_pred_2021)
	print("2021")
	print('MAE: %.3f' % mae_2021)
	print('MAPE: %.3f' % mape_2021)
	# plot expected vs actual
	pyplot.plot(forecast['ds'].values, y_true_2021, label='Actual')
	pyplot.plot(forecast['ds'].values, y_pred_2021, label='Predicted')
	pyplot.legend()
	pyplot.show()
 	# Save the data of 2021
 
	# define the period for which we want a prediction for 2020
	future = list()
	for i in range(1, 13):
		date = f"{2020}-{i}-01"
		future.append([date])
	future = DataFrame(future)
	future.columns = ['ds']
	future['ds'] = to_datetime(future['ds'],format="%Y/%m/%d")
	# use the model to make a forecast
	forecast = model.predict(future)
 
	# calculate MAE between expected and predicted values for 2020
	y_true_2020 = full_dataframe_transform['y'][-24:-12].values
	y_pred_2020 = forecast['yhat'].values
	mape_2020 = mean_absolute_percentage_error(y_true_2020, y_pred_2020)
	mae_2020 = mean_absolute_error(y_true_2020, y_pred_2020)
	print("2020")
	print('MAE: %.3f' % mae_2020)
	print('MAPE: %.3f' % mape_2020)
	# plot expected vs actual
	pyplot.plot(forecast['ds'].values, y_true_2020, label='Actual')
	pyplot.plot(forecast['ds'].values, y_pred_2020, label='Predicted')
	pyplot.legend()
	pyplot.show()
 	# Save the data of 2020
 
 	# define the period for which we want a prediction for 2019
	future = list()
	for i in range(1, 13):
		date = f"{2019}-{i}-01"
		future.append([date])
	future = DataFrame(future)
	future.columns = ['ds']
	future['ds'] = to_datetime(future['ds'],format="%Y/%m/%d")
	# use the model to make a forecast
	forecast = model.predict(future)
 
 	# calculate MAE between expected and predicted values for 2019
	y_test_2019 = full_dataframe_transform['y'][-36:-24].values
	y_pred_2019 = forecast['yhat'].values
	mape_2019 = mean_absolute_percentage_error(y_test_2019, y_pred_2019)
	mae_2019 = mean_absolute_error(y_test_2019, y_pred_2019)
	print("2019")
	print('MAE: %.3f' % mae_2019)
	print('MAPE: %.3f' % mape_2019)
 	# Save the data of 2019
	#print(y_pred_2019)
	#print(y_test_2019)
	# plot expected vs actual
	pyplot.plot(forecast['ds'].values, y_test_2019, label='Actual')
	pyplot.plot(forecast['ds'].values, y_pred_2019, label='Predicted')
	pyplot.legend()
	pyplot.show()
 


 	#export
	y_export = y_pred_2019
	y_export = np.append(y_export, y_pred_2020)
	y_export = np.append(y_export, y_pred_2021)

	#df_export_pred = pd.DataFrame(y_export, index = full_dataframe.index[-36:], columns=[f'{col_target}'])
	df_export_pred = pd.DataFrame(full_dataframe.index[-36:], columns=['Date'])
	df_export_pred['Depenes'] = y_export
	df_export_pred['Secteur'] = f'{col_target}'

	#df_export_pred_mae =pd.DataFrame([mae_2021,mape_2021,mae_2020,mape_2020,mae_2019,mape_2019],index=["mae_2021","mape_2021","mae_2020","mape_2020","mae_2019","mape_2019"],columns=[f'{col_target}'])
	df_export_pred_mae = pd.DataFrame([[2021,mae_2021,mape_2021],[2020,mae_2020,mape_2020],[2019,mae_2019,mape_2019]], columns=["Year","mae","mape"])
	df_export_pred_mae['Secteur'] = f'{col_target}'

	return df_export_pred,df_export_pred_mae

```
